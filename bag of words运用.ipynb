```python
import pandas as pd
text_ana = pd.read_csv('豆瓣热辣滚烫影评.csv', index_col = 0)
text_ana
```

```python
text = list(text_ana['评论概要'].dropna())
text
```

```python
import jieba
def word_segmentation(text):
    words = jieba.lcut(text)
    return ' '.join(words)
```

```python
text_ana['评论概要'] = text_ana['评论概要'].astype(str).apply(word_segmentation)
text_ana['评论概要'].head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
```

```python
X = vectorizer.fit_transform(text)
```

```python
print('词袋特征矩阵的形状：', X.shape)
```

```python
print('特征词列表：', vectorizer.get_feature_names_out())
```

```python
feature_words = vectorizer.get_feature_names_out()
```

```python
word_freq = dict(zip(feature_words, X.sum(axis = 0).A1))
```

```python
word_freq
```

```python
sorted(word_freq.items(),key = lambda x: x[1], reverse = True)
```

```python
wc = WordCloud(width = 1200,
              height = 800,
              background_color = 'white',
              font_path="C:/Windows/Fonts/MSYHL.ttc").generate_from_frequencies(word_freq)
```
